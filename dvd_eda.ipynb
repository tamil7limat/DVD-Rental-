```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
table1=pd.read_csv("C:/Users/admin/Downloads/dvdall_table3.csv")
table1
```

```python
table1.info()
```

```python
table1['filmcategory'].unique()
```

```python
import matplotlib.pyplot as plt

# Plotting the distribution of rental rates
plt.figure(figsize=(10, 6))
table1['rental_rate'].value_counts().sort_index().plot(kind='bar', color='lightblue', edgecolor='black')

# Adding title and labels
plt.title('Distribution of Rental Rates')
plt.xlabel('Rental Rate')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Plotting the distribution of film length
plt.figure(figsize=(10, 6))
plt.hist(table1['length'], bins=20, color='lightgreen', edgecolor='black')

# Adding title and labels
plt.title('Distribution of Film Length (Minutes)')
plt.xlabel('Film Length (minutes)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Count the number of films in each category
category_counts = table1['filmcategory'].value_counts()

# Plot the top film categories
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='salmon', edgecolor='black')

# Adding title and labels
plt.title('Top Film Categories')
plt.xlabel('Category')
plt.ylabel('Number of Films')

# Rotate category labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Grouping by customer_id to count the number of rentals
customer_rental = table1.groupby('customer_id').size().sort_values(ascending=False)

# Plot the number of rentals per customer
import matplotlib.pyplot as plt

plt.figure(figsize=(20,6))
customer_rental.head(80).plot(kind='bar', color='green', alpha=0.7)

# Adding title and labels
plt.title('Number of Rentals per Customer')
plt.xlabel('Customer ID')
plt.ylabel('Number of Rentals')

# Show the plot
plt.show()
```

```python
# Calculate the average rental rate by film category
avg_rental_rate_by_category = table1.groupby('filmcategory')['rental_rate'].mean()

# Plot the average rental rate by category
plt.figure(figsize=(12, 6))
avg_rental_rate_by_category.plot(kind='bar', color='purple', edgecolor='black')

# Adding title and labels
plt.title('Average Rental Rate by Film Category')
plt.xlabel('Film Category')
plt.ylabel('Average Rental Rate')

# Show the plot
plt.show()
```

```python
# Count the number of films each actor has starred in
actor_counts = table1['film_actor'].value_counts().head(10)  # Top 10 actors

# Plot the top 10 film actors
plt.figure(figsize=(12, 6))
actor_counts.plot(kind='bar', color='teal', edgecolor='black')

# Adding title and labels
plt.title('Top 10 Film Actors by Number of Films')
plt.xlabel('Actor')
plt.ylabel('Number of Films')

# Show the plot
plt.show()
```

```python
# Plotting the distribution of payments
plt.figure(figsize=(10, 6))
plt.hist(table1['payment'], bins=20, color='purple', edgecolor='black')

# Adding title and labels
plt.title('Distribution of Payments')
plt.xlabel('Payment Amount')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Data Preprocessing (Feature Engineering)
table1['film_age'] = 2024 - table1['release_year']  # Assuming current year is 2024
table1['rent_per_minute'] = table1['payment'] / table1['length']

# Rental Duration Categories (for better segmentation)
table1['rental_duration_category'] = pd.cut(table1['rental_duration'], bins=[0, 3, 7, 14, 30], labels=['Short', 'Medium', 'Long', 'Extended'])

# Univariate Analysis
plt.figure(figsize=(10, 6))
sns.histplot(table1['payment'], kde=True, color='skyblue', bins=30)
plt.title('Distribution of Payment Amount')
plt.xlabel('Payment')
plt.ylabel('Frequency')
plt.show()
```

```python

# Revenue by Category
category_revenue = table1.groupby('filmcategory')['payment'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
category_revenue.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Revenue by Film Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Rent per Minute vs. Rental Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=table1, x='rental_rate', y='rent_per_minute', color='purple')
plt.title('Rent per Minute vs. Rental Rate')
plt.xlabel('Rental Rate')
plt.ylabel('Rent per Minute')
plt.show()
```

```python
# Correlation Heatmap
corr_matrix = table1[['rental_rate', 'length', 'payment', 'rental_duration']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
